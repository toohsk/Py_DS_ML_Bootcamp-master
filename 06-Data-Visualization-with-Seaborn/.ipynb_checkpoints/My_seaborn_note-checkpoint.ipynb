```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
# Load tips dataset as pandas.DataFrame
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

# distribution plots

```python
sns.distplot(tips['total_bill'])
# you can remove kde line by sns.distplot(tips['total_bill'], kde=False)
```

```python
sns.distplot(tips['total_bill'], bins=40)
```

```python
# you can show two data's distribution plot and scattor plot
#sns.joinplot(x=, y=, data=)

sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
# to show scattor plot with hexagon
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
# to show with regression line
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

```python
sns.pairplot(tips)
sns.pairplot(tips, hue='sex')
```

# Categorical plots

```python
import numpy as np
```

```python
# you can plot categorical data as a bar char with sns.barplot(x=, y=, data=)
sns.barplot(x='sex', y='total_bill', data=tips)
```

```python
# you can set your original aggrigator as estimator
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

```python
# this will plot the count distribution, and also you can only define single catogorical column
sns.countplot(x='sex', data=tips)
```

```python
# Box plot
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```

```python
# Violin plot
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)

# When communicating with other data scientists, maybe violin plot would be good plot, because it has more information rather than box plot.
# But, when you are communicating with CEO or other managers, maybe box plot would be good, because it's simple and familor to those persons.
```

```python
sns.stripplot(x='day', y='total_bill', data=tips)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex', dodge=True)
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips, color='black')
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
sns.factorplot(x='day', y='total_bill', data=tips, kind='bar')
```

```python
sns.factorplot(x='day', y='total_bill', data=tips, kind='violin')
```

# Matrix Plot

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
sns.heatmap(tips.corr(), annot=True)
```

```python
fp = flights.pivot_table(index='month', columns='year', values='passengers')
```

```python
sns.heatmap(fp)
```

```python
# to add seperations
sns.heatmap(fp, linecolor='white', linewidths=2)
```

```python
sns.clustermap(fp)
```

```python
sns.clustermap(fp, standard_scale=1)
```

# Grids

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.pairplot(iris)
```

```python
import matplotlib.pyplot as plt
```

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
#fg = sns.FacetGrid(data=tips, col=, row=)

fg = sns.FacetGrid(data=tips, col='time', row='smoker')
fg.map(sns.distplot, 'total_bill')
```

```python
fg = sns.FacetGrid(data=tips, col='time', row='smoker')
fg.map(plt.scatter, 'total_bill', 'tip')
```

# Regression plots

```python
# lmplot stands for Linear Model Plot
# axis need to be numerical data
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o', 'v'])
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', hue='sex', aspect=0.5, height=8)
```
