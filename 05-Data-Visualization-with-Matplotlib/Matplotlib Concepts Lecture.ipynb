```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y, '-r')
```

```python
plt.plot(x, y)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x,y,'r')

plt.subplot(1, 2, 2)
plt.plot(y,x,'b')
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes_1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes_2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes_1.plot(x, y)
axes_1.set_title('LARGER PLOT')

axes_2.plot(y, x)
axes_2.set_title('SMALLER PLOT')
```

```python
fig, axes = plt.subplots()

axes.plot(x, y)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y)
axes[0].set_title('Fig 1')

axes[1].plot(y, x)
axes[1].set_title('Fig 2')
```

# Figure size and DPI

```python
fig =  plt.figure(figsize=(3,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x, y)
```

```python
fig.savefig('matplot_example.png')
```

```python
fig = plt.figure(figsize=(3,2))

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label='X Squared')
ax.plot(x, x**3, label='X Cubed')

ax.legend()
# You can define the location of lengend box with loc=(left, bottom)
# ax.legend(loc=(0.05, 0.5))
```

```python
# Plot Appearance
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, color='green', linewidth=3, alpha=0.2)
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, color='green', lw=3, linestyle='--')

# you can use ls instead of linestyle
# ax.plot(x, y, color='green', lw=3, linestyle='--')
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, color='green', lw=1, ls='-', marker='o', markersize=10, markerfacecolor='blue', markeredgewidth=3)

# you can use ls instead of linestyle
# ax.plot(x, y, color='green', lw=3, linestyle='--')
```
